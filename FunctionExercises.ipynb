```python
def multiplication_by_2(x):
    return x * 2

print(multiplication_by_2(2))
```

```python
def division_by_2(x):
    return float(x/2)

print(division_by_2(2))
```

```python
def exponentiation_exp_2(x) :
    print(str(x), "Raised to the power of 2:")
    result = x ** 2
    return result
```

```python
exponentiation_exp_2(2)
```

```python
def plus_five(x):
    return x + 5

def m_by_3(x):
    return plus_five(x) * 3
    
m_by_3(5)
```

```python
def compare_the_two(f,s):
    if f > s:
        print("Greater")
    elif s > f:
        print("Less")
    else:
        print("Equal")
```

```python
compare_the_two(4,6)
```

```python
def distance_from_zero(x):
    if type(x) != int:
        print("Not possible")
    else:
        return abs(x)
```
