```python
prices = {
    "box_of_spaghetti" : 4,
    "lasagna"  : 5,
    "hamburger" : 2
   }
quantity = {
    "box_of_spaghetti" : 6,
    "lasagna"  : 10,
    "hamburger" : 0
    }

money_spent = 0
for meal in prices:
    if prices[meal] >= 5:
        money_spent = money_spent + prices[meal] * quantity[meal]
print(money_spent)
```
