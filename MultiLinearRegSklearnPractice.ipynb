# Create Multiple Linear Regression

# Declare the dependant and independant variable

```python
x = data[['SAT', 'Rand 1,2,3']]
y = data['GPA']
```

# Regression iteself

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

# Calculating the R-squared

```python
reg.score(x,y)
```

```python
#Adjusted R-squared is more appropiate for Multilinear Reg
```

# Formula for Adjusted R^2
$ R^2_{adj.} = 1 - (1-R^2)*\frac{n-1}{n-p-1}$

```python
# n = 84
# p = 2
x.shape
```

```python
r2 = reg.score(x,y)
n = x.shape[0]
p = x.shape[1]

adjusted_r2 = 1 - (1-r2)*(n-1)/(n-p-1)
adjusted_r2
```
